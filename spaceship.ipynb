```python
# Importando o pandas
import pandas as pd
```

```python
# Lendo o df de treino
df_train = pd.read_csv('train.csv')
```

```python
df_train.head()
```

```python
# Verificando a quantidade de valores vazios para cada coluna
df_train.isnull().sum().sort_values(ascending=False).head(13)
```

```python
# Vendo a quantidade de valores na coluna CryoSleep
df_train['CryoSleep'].value_counts()
```

```python
# Substituindo os nulos dessa coluna pela maior quantidade
df_train['CryoSleep'] = df_train['CryoSleep'].fillna(False)
```

```python
# Vendo a quantidade de valores na coluna VIP
df_train['VIP'].value_counts()
```

```python
# Substituindo os nulos dessa coluna pela maior quantidade
df_train['VIP'] = df_train['VIP'].fillna(False)
```

```python
# Vendo a moda da coluna Age
age_mode = df_train['Age'].mode()
age_mode
```

```python
# Substituindo os nulos dessa coluna pela moda
df_train['Age'] = df_train['Age'].fillna(age_mode.values[0])
```

```python
# Vendo a quantidade de valores na coluna HomePlanet
df_train['HomePlanet'].value_counts()
```

```python
# Substituindo os nulos pelo valor que mais aparece
df_train['HomePlanet'] = df_train['HomePlanet'].fillna('Earth')
```

```python
# Vendo a quantidade de valores na coluna Destination
df_train['Destination'].value_counts()
```

```python
# Substituindo os nulos pelo valor que mais aparece
df_train['Destination'] = df_train['Destination'].fillna('TRAPPIST-1e')
```

```python
# Substituindo todos os outros valores numéricos pela sua média de sua coluna
vr_mean = df_train['VRDeck'].mean()
food_mean = df_train['FoodCourt'].mean()
spa_mean = df_train['Spa'].mean()
shop_mean = df_train['ShoppingMall'].mean()
room_mean = df_train['RoomService'].mean()

df_train['VRDeck'] = df_train['VRDeck'].fillna(vr_mean)
df_train['FoodCourt'] = df_train['FoodCourt'].fillna(food_mean)
df_train['Spa'] = df_train['Spa'].fillna(spa_mean)
df_train['ShoppingMall'] = df_train['ShoppingMall'].fillna(shop_mean)
df_train['RoomService'] = df_train['RoomService'].fillna(room_mean)
```

```python
# Vendo que todos os valores nulos foram preenchidos
df_train.isnull().sum().sort_values(ascending=False).head(13)
```

```python
# Transformando os valores True em 1 e False em 0, e as strings com no máximo 3 valroes diferentes em 0, 1 e 2
def vip(VIP):
    if VIP == True:
        return 1
    else:
        return 0

def transport(Transported):
    if Transported == True:
        return 1
    else:
        return 0

def CryoSleep2(CryoSleep):
    if CryoSleep == True:
        return 1
    else:
        return 0

def HomePlanet2(HomePlanet):
    if HomePlanet == 'Earth':
        return 0
    elif HomePlanet == 'Europa':
        return 1
    else:
        return 2

def Destination2(Destination):
    if Destination == 'TRAPPIST-1e':
        return 0
    elif Destination == 'PSO J318.5-22':
        return 1
    else:
        return 2

df_train['VIP'] = df_train['VIP'].apply(vip)
df_train['Transported'] = df_train['Transported'].apply(transport)
df_train['CryoSleep'] = df_train['CryoSleep'].apply(CryoSleep2)
df_train['HomePlanet'] = df_train['HomePlanet'].apply(HomePlanet2)
df_train['Destination'] = df_train['Destination'].apply(Destination2)
```

```python
# Excluindo do df as colunas com strings que serão inúteis para a análise
df_train = df_train.drop(['Name', 'Cabin'], axis=1)
```

```python
df_train.head()
```

```python
df_train.info()
```

```python
# Importando o train_test_split
from sklearn.model_selection import train_test_split
```

```python
# Separando o X e o y
X = df_train.drop(['PassengerId', 'Transported'], axis=1)
y = df_train.Transported
```

```python
# Separando em dados de treino e teste
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
# Importando a árvore de decisão
from sklearn import tree
```

```python
# Aplicando a árvore de decisão aos dados de teste e fazendo a predição
clf_ac = tree.DecisionTreeClassifier(random_state=42)
clf_ac = clf_ac.fit(X_train, y_train)
y_pred_ac = clf_ac.predict(X_val)
```

```python
# Importando o KNN
from sklearn.neighbors import KNeighborsClassifier
```

```python
# Aplicando o KNN aos dados de treino e fazendo a predição
clf_knn = KNeighborsClassifier(n_neighbors=3)
clf_knn = clf_knn.fit(X_train, y_train)
y_pred_knn = clf_knn.predict(X_val)
```

```python
# Importando a regressão logística
from sklearn.linear_model import LogisticRegression
```

```python
# Aplicando a regressão logística e fazendo a predição
clf_rl = LogisticRegression(random_state=42)
clf_rl = clf_rl.fit(X_train, y_train)
y_pred_rl = clf_rl.predict(X_val)
```

```python
# Importando a acurácia
from sklearn.metrics import accuracy_score
```

```python
# Vendo a acurácia da árvore de decisão
accuracy_score(y_val, y_pred_ac)
```

```python
# Vendo a acurácia do KNN
accuracy_score(y_val, y_pred_knn)
```

```python
# Vendo a acurácia da regressão logística
accuracy_score(y_val, y_pred_rl)
```

```python
# Importando a matriz de confusão
from sklearn.metrics import confusion_matrix
```

```python
# Matriz de confusão para a árvore de decisão
confusion_matrix(y_val, y_pred_ac)
```

```python
# Matriz de confusão para o KNN
confusion_matrix(y_val, y_pred_knn)
```

```python
# # Matriz de confusão para a regressão logística
confusion_matrix(y_val, y_pred_rl)
```

Após essa análise, conseguimos ver que a regressão logística foi responsável pela melhor acurácia, agora tentaremos normalizar os dados para ver se há uma melhora em alguma acurácia
